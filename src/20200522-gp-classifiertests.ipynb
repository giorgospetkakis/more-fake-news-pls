```python
Xtrain, Xtest, ytrain, ytest, IDtrain, IDtest = train_test_split(X, y, IDs, test_size=0.33, random_state=1)
model = AdaBoostClassifier(RandomForestClassifier(max_depth=1), n_estimators=200, learning_rate=0.26, random_state=0)
model.fit(Xtrain, ytrain)
pred= model.predict(Xtest).astype(int)
wrong = []
for i in range(len(ytest)):
    if ytest[i] != pred[i]:
        wrong.append(IDtest[i])
```

```python
right = list(set(IDtest).difference(set(wrong)))
```

```python
Authors = data.get_processed_data()
```

```python
w1_ENTS = []
w1_ENTS_tags = []
w0_ENTS = []
w0_ENTS_tags = []
for idtag in wrong:
    for tweet in Authors[idtag].ents:
        for entity in tweet:
            if Authors[idtag].truth == 1:
                w1_ENTS.append(entity[0])
                w1_ENTS_tags.append(entity[1])
            else:
                w0_ENTS.append(entity[0])
                w0_ENTS_tags.append(entity[1])
r1_ENTS = []
r1_ENTS_tags = []
r0_ENTS = []
r0_ENTS_tags = []
for idtag in right:
    for tweet in Authors[idtag].ents:
        for entity in tweet:
            if Authors[idtag].truth == 1:
                r1_ENTS.append(entity[0])
                r1_ENTS_tags.append(entity[1])
            else:
                r0_ENTS.append(entity[0])
                r0_ENTS_tags.append(entity[1])
```

```python
# Incorrectly labelled "true" news and the correctly labeled

fig,axes = plt.subplots(nrows=1, ncols=2)
fig.suptitle("True news",size=30)
fig.set_size_inches(12,12)

# Iterate through sigmas and plot each version

axes[0].set_title("NER Tags")
axes[0].hist(r1_ENTS_tags, label='right')
axes[0].hist(w1_ENTS_tags, label='wrong')
axes[0].tick_params(labelrotation=90)
axes[0].legend()

axes[1].set_title("Entities")
r1 = hist(r1_ENTS, label='right')
w1 = hist(w1_ENTS, label='wrong')
axes[1].tick_params(labelrotation=90)
axes[1].legend()
    
plt.show()
```

```python
r1_counts = {}
w1_counts = {}
for entity in set(r1_ENTS):
    r1_counts[entity] = 0
    w1_counts[entity] = 0
for entity in set(w1_ENTS):
    w1_counts[entity] = 0
    r1_counts[entity] = 0
    
for item in r1_ENTS:
    r1_counts[item] += 1
for item in w1_ENTS:
    w1_counts[item] += 1
    
correct1 = sorted(r1_counts, key=r1_counts.get, reverse=True)[:30]
wrong1 = sorted(w1_counts, key=w1_counts.get, reverse=True)[:30]
r1_subset_of_w = {key: r1_counts[key] for key in wrong1}
w1_subset_of_w = {key: w1_counts[key] for key in wrong1}
r1_subset_of_c = {key: r1_counts[key] for key in correct1}
w1_subset_of_c = {key: w1_counts[key] for key in correct1}
```
